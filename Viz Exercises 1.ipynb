### Step 1. Import the necessary libraries:

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

### Step 2. Import the dataset in the PWD

```python
pd.read_csv("tips.csv")
```

### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv("tips.csv")
```

### Step 4. Delete the Unnamed 0 column

```python
del tips['Unnamed: 0']
```

### Step 5. Plot the total_bill column histogram

```python
sb.set(style="darkgrid") 
```

```python
x = tips["total_bill"]
sb.distplot(x, kde = True, hist = True, rug= False, bins= 40)
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
sb.jointplot(x = "total_bill",y = "tip",data = tips)
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
sb.relplot(x = 'total_bill', y = 'tip', style = 'size', data = tips)
```

### Step 8. Present the relationship between days and total_bill value

```python
sb.catplot(x = 'day', y = 'total_bill', data = tips)
```

```python
# Interpretation 
# The total bill paid on saturday is more,compare to Sunday, Thursday and Friday.
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sb.relplot(x="tip", y="day", hue="sex", data=tips);
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sb.catplot(x = 'day', y = 'total_bill', kind = 'box', data = tips, hue = 'time')
```
